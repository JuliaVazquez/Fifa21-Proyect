# Standardizing header

```python
col1 = list(map(lambda el : el.lower(),fifa21))
fifa21.columns = col1

col = fifa21.columns.str.replace(" ","_")
fifa21.columns = col

fifa21.columns
```

# Data Cleaning

### Dropping first columns with unique values

```python
fifa21 = fifa21.drop(['id', 'name'], axis=1)
```

### Checking for columns that are the sum of others and dropping them

ls-gk --> data about best fit position, but we also have it in the bp column so we choose to drop them

goalkeeper = gk_driving - gk_reflexes

defending = marking - sliding_tackle

mentality = aggression - penalties

power = shot_power - long_shots

movement = acceleration - balance

skill = dribbling - ball_control

attacking = crossing - volleys

base_stats = pac-phy

```python
def dropping_cols(df,indexes1,indexes2):
   
    for i in range (len(indexes1)):
        cols = list(df.columns)
        min_index = cols.index(indexes1[i])
        max_index = cols.index(indexes2[i])
        df = df.drop(cols[min_index:max_index+1],axis=1)
    return df
```

```python
first = ['ls','gk_diving','marking','aggression','shot_power','acceleration','dribbling','crossing','pac']
last = ['gk','gk_reflexes','sliding_tackle','penalties','long_shots','balance','ball_control','volleys','phy']
fifa21 = dropping_cols(fifa21,first,last)
```

### Adjusting weight and height

```python
col = fifa21['weight'].str.replace("lbs","")
fifa21['weight'] = col
fifa21['weight'] = pd.to_numeric(fifa21['weight'], errors='coerce')
```

```python
fifa21['height'] = fifa21['height'].str.replace("'", '+')
fifa21['height'] = fifa21['height'].str.replace('"', '/12')
fifa21['height'] = round(fifa21['height'].apply(eval),2)
```

### Checking columns that must be numerical but are categorical and dealing with them

Removing € , M , K

```python
def replace_K (df, column):
    for col in column:
        df[col] = df[col].str.replace('€','').str.replace('K','*1000').str.replace('M','*1000000').apply(eval)
    return df
```

```python
fifa21 = replace_K(fifa21, ['value','wage','release_clause','hits'])
```

Removing ★

```python
def replace_stars(df, column):
    for col in column:
        df[col] = pd.to_numeric((df[col].str.replace('★','')), errors='coerce')
    return df
```

```python
stars = ['w/f','sm', 'ir']
fifa21 = replace_stars(fifa21, stars)
```

# X y Split

```python
y = fifa21['ova']
X = fifa21.drop(['ova'],axis=1)

X_num = X.select_dtypes(np.number)
X_cat = X.select_dtypes(object)
```

### Normalize numeric values

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
MinMaxtransformer = MinMaxScaler().fit(X_num)
X_normalized = MinMaxtransformer.transform(X_num)
X_normalized = pd.DataFrame(X_normalized,columns=X_num.columns)
X_normalized.head()
```

### Encode categorical values and concat new_X

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
def one_hot_encoder(col):
    col_df = pd.DataFrame(col)
    one_hot = OneHotEncoder(drop='first').fit(col_df)
    encoded_array = one_hot.transform(col_df).toarray()
    print(encoded_array)
    col_names = col.unique()
    col_names = np.delete(col_names, [0])
    print(col_names)
    df_encoded = pd.DataFrame(encoded_array, columns=col_names)
    return df_encoded
```

```python
new_X = X_normalized

for name in X_cat.columns:
    encoded_X = one_hot_encoder(X_cat[name])
    new_X = pd.concat([new_X, encoded_X], axis=1)

new_X
```

# Train-test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(new_X, y, test_size=0.2, random_state=42)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Train/fit our model

```python
from sklearn import linear_model
```

```python
lm = linear_model.LinearRegression()
print(lm.fit(X_train,y_train))
```

# Testing our model

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
predictions_train = lm.predict(X_train)
print(round(r2_score(y_train, predictions_train),2))
```

```python
predictions_test = lm.predict(X_test)
print(round(r2_score(y_test, predictions_test),2))
```

```python
mse = mean_squared_error(y_test,predictions_test)
rmse = np.sqrt(mean_squared_error(y_test,predictions_test))

print(round(mse,2))
print(round(rmse,2))
```

```python
print(round(mean_absolute_error(y_train,predictions_train),2))
print(round(mean_absolute_error(y_test,predictions_test),2))
```

```python
def cleaner (df):
    
    # Standardizing header
    col1 = list(map(lambda el : el.lower(),df))
    df.columns = col1
    col = df.columns.str.replace(" ","_")
    df.columns = col
    
    ### Dropping first columns with unique values
    df = df.drop(['id', 'name'], axis=1)
    
    ### Dropping columns that are sum of others
    first = ['ls','gk_diving','marking','aggression','shot_power','acceleration','dribbling','crossing','pac']
    last = ['gk','gk_reflexes','sliding_tackle','penalties','long_shots','balance','ball_control','volleys','phy']
    df = dropping_cols(df,first,last)
    
    ### Adjusting weight and height
    col = df['weight'].str.replace("lbs","")
    df['weight'] = col
    df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
    
    df['height'] = df['height'].str.replace("'", '+')
    df['height'] = df['height'].str.replace('"', '/12')
    df['height'] = round(df['height'].apply(eval),2)
    
    ###Removing € , M , K
    df = replace_K(df, ['value','wage','release_clause','hits'])
    
    ###Removing ★
    stars = ['w/f','sm', 'ir']
    df = replace_stars(df, stars)
    
    ### Dropping redundant values or that give us less information due to its excess of uniques
    df = df.drop(df[['position','joined', 'nationality', 'club']],axis=1)
    
    ### Changing columns that have categorical values easy convertibles to numeric
    df['a/w'] = pd.to_numeric((df['a/w'].replace({'Low': '1', 'Medium' : '2', 'High':'3'})),errors='coerce')
    df['d/w'] = pd.to_numeric((df['d/w'].replace({'Low': '1', 'Medium' : '2', 'High':'3'})),errors='coerce')
    
    ### Checking for NaN and strange values and dropping redundat columns
    df = df[df['wage']>0]
    df = df.drop(['loan_date_end','team_&_contract','contract'], axis=1)
    df = df.dropna()
    
    ### Removing columns with high correlation that are redundant
    df = df.drop(['total_stats','release_clause','attacking'],axis=1)
    
    ### X y Split
    y = df['ova']
    X = df.drop(['ova'],axis=1)

    X_num = X.select_dtypes(np.number)
    X_cat = X.select_dtypes(object)
    
    ### Normalize numeric values
    MinMaxtransformer = MinMaxScaler().fit(X_num)
    X_normalized = MinMaxtransformer.transform(X_num)
    X_normalized = pd.DataFrame(X_normalized,columns=X_num.columns)
    X_normalized.head()
    
    ### Encode categorical values and concat new_X
    new_X = X_normalized

    for name in X_cat.columns:
        encoded_X = one_hot_encoder(X_cat[name])
        new_X = pd.concat([new_X, encoded_X], axis=1)

    return new_X,y    
    
```

```python
new_X_test,new_y_test = cleaner(new_test_df)
print(new_X_test.shape)
print(new_y_test.shape)
```

```python
new_predictions_test = lm.predict(new_X_test)
print(round(r2_score(new_y_test, new_predictions_test),2))
```

```python
new_mse = mean_squared_error(new_y_test,new_predictions_test)
new_rmse = np.sqrt(mean_squared_error(new_y_test,new_predictions_test))

print(round(new_mse,2))
print(round(new_rmse,2))
```
